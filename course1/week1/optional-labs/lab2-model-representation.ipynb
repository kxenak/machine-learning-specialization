```python
import numpy as np
import matplotlib.pyplot as plt
```

We will create a house prediction model. We will only use two data points to keep the model simple. 

```python
# x_train is the input variable (size of houses in 1000 sq feet)
# y_train is the output variable (price of house in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m is the number of training examples
m = x_train.shape[0]
print(f"The number of training examples = {m}")
```

```python
# printing the first training example
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"First training example = (x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker = "x", c = "r")
plt.title("Housing Prices")
plt.xlabel("Size (In 1000 sq feet)")
plt.ylabel("Price (In 1000s of Dollars)")
plt.show()
```

```python
# checking with different parameters w and b
w = 100
b = 100
print(f"w = {w}")
print(f"b = {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear regression model.
    Arg:
        x (ndarray (m,)): Data, m examples
        w, b (scalar): model parameters
    Returns
        f_wb (ndarray (m,)): target variables
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
print(compute_model_output(x_train, 100, 100))
```

```python
# Plotting the predicted values along with the actual values.
predicted_values = compute_model_output(x_train, 200, 100)
plt.plot(x_train, predicted_values, c = "b", label = "Predicted Values")
plt.scatter(x_train, y_train, marker = "x", c = "r")
plt.title("Housing Prices")
plt.xlabel("Size (In 1000s of square feet)")
plt.ylabel("Price (In the 1000s of dollars.)")
plt.show()
```

```python
# making an original prediction
w = 200
b = 100
x_i = 1.2
original_prediction = w * x_i + b
print(f"Original prediction = {original_prediction:.0f}$")
```
